# Step 1: Import Cleaned Dataset into SQLite

### Objective
Import the cleaned Superstore dataset into a SQLite database named **superstore.db**. The imported table will be used to perform SQL queries for data extraction and analysis.

```python
import pandas as pd
import sqlite3
```

```python
# Load the cleaned dataset

df = pd.read_csv("../data/cleaned_superstore.csv")

# Display first five rows
df.head()
```

```python
# Connect to SQLite Database

conn = sqlite3.connect("../database/superstore.db")

print("Connected Successfully!")
```

```python
# Import dataset into SQLite

df.to_sql(
    "superstore_data",
    conn,
    if_exists="replace",
    index=False
)

print("Dataset Imported Successfully!")
```

```python
# Check table names

query = """
SELECT name
FROM sqlite_master
WHERE type='table';
"""

pd.read_sql(query, conn)
```

```python
# Display first five records

query = """
SELECT *
FROM superstore_data
LIMIT 5;
"""

pd.read_sql(query, conn)
```

```python
# Check column names

query = """
PRAGMA table_info(superstore_data);
"""

pd.read_sql(query, conn)
```

# Step 3: Aggregate Functions, GROUP BY and HAVING

### Objective

This step demonstrates how SQL aggregate functions are used to summarize data and how `GROUP BY` and `HAVING` help analyze data by different categories.

```python
query = """
SELECT
    SUM(Sales) AS Total_Sales
FROM superstore_data;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    AVG(Profit) AS Average_Profit
FROM superstore_data;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    COUNT(*) AS Total_Orders
FROM superstore_data;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    Category,
    SUM(Sales) AS Total_Sales
FROM superstore_data
GROUP BY Category;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    Category,
    SUM(Profit) AS Total_Profit
FROM superstore_data
GROUP BY Category;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    Category,
    SUM(Sales) AS Total_Sales
FROM superstore_data
GROUP BY Category
ORDER BY Total_Sales DESC;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    Category,
    SUM(Sales) AS Total_Sales
FROM superstore_data
GROUP BY Category
HAVING SUM(Sales) > 500000;
"""

pd.read_sql(query, conn)
```

## Conclusion

The following SQL concepts were successfully implemented:

- SUM()
- AVG()
- COUNT()
- GROUP BY
- HAVING

These operations are commonly used in business reporting and dashboard creation to summarize and analyze data.

# Step 4: Advanced SQL

### Objective

In this step, advanced SQL concepts such as Subqueries, Common Table Expressions (CTEs), Window Functions, and Views are implemented to perform advanced data analysis.

```python
query = """
SELECT
    "Customer Name",
    Sales
FROM superstore_data
WHERE Sales >
(
    SELECT AVG(Sales)
    FROM superstore_data
);
"""

pd.read_sql(query, conn)
```

```python
query = """
WITH CategorySales AS
(
    SELECT
        Category,
        SUM(Sales) AS Total_Sales
    FROM superstore_data
    GROUP BY Category
)

SELECT *
FROM CategorySales;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    "Customer Name",
    Category,
    Sales,
    ROW_NUMBER() OVER (ORDER BY Sales DESC) AS Row_Num
FROM superstore_data;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    "Customer Name",
    Sales,
    RANK() OVER (ORDER BY Sales DESC) AS Sales_Rank
FROM superstore_data;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    "Order Date",
    Sales,
    LAG(Sales) OVER (ORDER BY "Order Date") AS Previous_Sales
FROM superstore_data;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    "Order Date",
    Sales,
    LEAD(Sales) OVER (ORDER BY "Order Date") AS Next_Sales
FROM superstore_data;
"""

pd.read_sql(query, conn)
```

```python
query = """
CREATE VIEW IF NOT EXISTS Category_Summary AS

SELECT
    Category,
    SUM(Sales) AS Total_Sales,
    SUM(Profit) AS Total_Profit

FROM superstore_data

GROUP BY Category;
"""

conn.execute(query)

print("View Created Successfully!")
```

```python
query = """
SELECT *
FROM Category_Summary;
"""

pd.read_sql(query, conn)
```

## Conclusion

The following advanced SQL concepts were successfully implemented:

- Subqueries
- Common Table Expressions (CTEs)
- ROW_NUMBER()
- RANK()
- LAG()
- LEAD()
- SQL Views

These techniques are widely used in business intelligence, reporting, and data analytics projects.

# Step 5: Business Insights using SQL

## Objective

In this step, SQL queries are executed to answer real-world business questions using the Superstore dataset.

```python
import sqlite3
import pandas as pd

# Reconnect to the SQLite database
conn = sqlite3.connect("../database/superstore.db")

print("Database Connected Successfully!")
```

## Question 1
What is the total sales generated by the company?

```python
query = """
SELECT
SUM(Sales) AS Total_Sales
FROM superstore_data;
"""

pd.read_sql(query, conn)
```

## Question 2:
 What is the total profit earned by the company?

```python
query = """
SELECT
SUM(Profit) AS Total_Profit
FROM superstore_data;
"""

pd.read_sql(query, conn)
```

## Question 3
 Which product category generates the highest sales?

```python
query = """
SELECT
Category,
SUM(Sales) AS Total_Sales
FROM superstore_data
GROUP BY Category
ORDER BY Total_Sales DESC;
"""

pd.read_sql(query, conn)
```

## Question 4

 Which region contributes the highest sales?

```python
query = """
SELECT
Region,
SUM(Sales) AS Total_Sales
FROM superstore_data
GROUP BY Region
ORDER BY Total_Sales DESC;
"""

pd.read_sql(query, conn)
```

## Question 5:
Who are the top 10 customers based on total sales?

```python
query = """
SELECT
"Customer Name",
SUM(Sales) AS Total_Sales
FROM superstore_data
GROUP BY "Customer Name"
ORDER BY Total_Sales DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

## Question 6

 Which are the top 10 best-selling products based on sales?

```python
query = """
SELECT
"Product Name",
SUM(Sales) AS Total_Sales
FROM superstore_data
GROUP BY "Product Name"
ORDER BY Total_Sales DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

## Question 7
Which customer segment earns the highest profit?

```python
query = """
SELECT
Segment,
SUM(Profit) AS Total_Profit
FROM superstore_data
GROUP BY Segment
ORDER BY Total_Profit DESC;
"""

pd.read_sql(query, conn)
```

## Question 9
 Which are the top 5 states with the highest sales?

```python
query = """
SELECT
State,
SUM(Sales) AS Total_Sales
FROM superstore_data
GROUP BY State
ORDER BY Total_Sales DESC
LIMIT 5;
"""

pd.read_sql(query, conn)
```

## Question 10
 What is the average discount provided for each product category?

```python
query = """
SELECT
Category,
AVG(Discount) AS Average_Discount
FROM superstore_data
GROUP BY Category;
"""

pd.read_sql(query, conn)
```
