## Day 14: Python Visualization Basics

### Objective
Create visualizations to understand sales, profit, category, and regional performance using Python.

### Tools Used
- Pandas
- Matplotlib
- Seaborn
- Plotly

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("../processed/cleaned_superstore.csv")
df.head()
```

## Sales Distribution
This histogram shows how sales how values are distributed across all transactions

```python
plt.figure(figsize=(5,2))
plt.hist(df["Sales"],bins=30)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("frequency")
plt.show()
```

## Sales by Category
compare total sales generated by each product category

```python
category_sales=df.groupby("Category")["Sales"].sum()
plt.figure(figsize=(6,3))
category_sales.plot(kind="bar")
plt.title("Total Sales by Category")
plt.ylabel("Sales")
plt.show()
```

## Sales by Region
Visualize region sales performance.

```python
reg_sales=df.groupby("Region")["Sales"].sum()
plt.figure(figsize=(5,2))
reg_sales.plot(kind="bar")
plt.title("Sales by region")
plt.ylabel("Sales")
plt.show()
```

## Profit By Category
Analyze profitability across product categories

```python
profit_Cat=df.groupby("Category")["Profit"].sum()
plt.figure(figsize=(5,2))
profit_Cat.plot(kind="bar")
plt.title("Profit by category")
plt.ylabel("Profit")
plt.show()
```

## Top 10 States by Sales

Identify the highest revenue-generating states.

```python
top_states = (
    df.groupby("State")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(8,4))
top_states.plot(kind="bar")

plt.title("Top 10 States by Sales")
plt.ylabel("Sales")

plt.show()
```

## Correlation Heatmap
Analyze relationships between  numerical variables

```python
plt.figure(figsize=(6,2))
sns.heatmap(df[["Sales","Profit","Quantity","Discount"]].corr(),annot=True)
plt.title("Correlation Heatmap")
```

## Sales vs Profit
Understand the relationship between sales and profit

```python
plt.figure(figsize=(6,3))

sns.scatterplot(
    data=df,
    x="Sales",
    y="Profit"
)

plt.title("Sales vs Profit")

plt.show()
```

## Profit Distribution by Category


```python
plt.figure(figsize=(5,2))

sns.boxplot(
    data=df,
    x="Category",
    y="Profit"
)

plt.title("Profit Distribution by Category")

plt.show()
```

```python
# Pairplot of important numerical columns
sns.pairplot(
    df[["Sales", "Profit", "Quantity", "Discount"]]
)

plt.show()
```

```python
# Boxen Plot for Profit by Category

plt.figure(figsize=(8, 5))

sns.boxenplot(
    data=df,
    x="Category",
    y="Profit"
)

plt.title("Profit Distribution by Category")
plt.xlabel("Category")
plt.ylabel("Profit")

plt.show()
```

```python
import plotly.express as px

# Total Sales by Category
category_sales = df.groupby("Category", as_index=False)["Sales"].sum()

fig = px.bar(
    category_sales,
    x="Category",
    y="Sales",
    title="Total Sales by Category",
    color="Category",
    text_auto=True
)

fig.show()
```

### Observation

- Technology contributes the highest sales.
- Furniture contributes moderate sales.
- Office Supplies also contributes significantly to total sales.

### Conclusion

The interactive Plotly chart allows users to explore sales data dynamically by hovering, zooming, and interacting with the visualization. This improves data exploration compared to static charts.
